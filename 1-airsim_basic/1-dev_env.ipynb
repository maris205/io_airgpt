# 1.1 Development Environment Setup

## 1 Basic Requirements
- Hardware: 16GB RAM, a GPU is recommended, at least 50GB free disk space.
- Software: Windows 11 is recommended.
- For macOS or Linux, you will need to compile AirSim from source.

## 4 Python Development Environment
Use conda to create a Python 3.10 environment called airsim_agent.

Then open it in PyCharm and install JupyterLab.

```
conda create -n airsim_agent python=3.10

conda activate airsim_agent

pip install jupyterlab

jupyter lab
```

Then download the course code:

git clone https://github.com/maris205/airsim_agent

Note: Due to version conflicts between AirSim's tornado dependency and JupyterLab, we do not install AirSim via pip. Instead, we directly import the locally downloaded package:
