# 2.3 Introduction to LLM Prompt Engineering

### **Programming in the LLM Era == Writing Prompts**


Prompt Engineering is a key technique for systematically optimizing AI interactions. By designing, debugging, and optimizing prompts, it precisely guides Large Language Models (LLMs) to produce expected outputs. Currently, most applications are accomplished through prompts. Here are the core points:

---

### **I. Definition and Core Objectives**
1. **Definition**
   Making LLMs understand human intent. Prompt engineering transforms vague user requirements into structured instructions, controlling model behavior through input content to reduce output randomness and unlock the model's latent capabilities. Its essence is building a standardized "human-AI collaboration" interaction flow, not just a questioning technique.

2. **Core Objectives**

- **Intent alignment**: Transform human needs into model-understandable expressions (e.g., through role-playing frameworks like "You are a senior lawyer").
- **Output control**: Reduce uncertainty through explicit instructions (e.g., format, length) such as requiring Markdown table output.
- **Capability activation**: Unlock the model's advanced reasoning and creative abilities (e.g., using Chain-of-Thought (CoT) techniques to guide complex problem solving).

---

### **II. Key Technical Elements**
1. **Structured Design**
   - **Modular construction**: Break prompts into modules like role, task, and constraints (e.g., "role + background + output requirements" framework).
   - **Parameter tuning**: Explicitly control model parameters (e.g., `temperature` to adjust output randomness).
   - **Multi-model collaboration**: Use multiple models in relay for complex tasks (e.g., GPT for copywriting, then Stable Diffusion for images).

2. **Optimization Methods**
   - **Iterative testing**: Optimize prompts through multiple rounds of debugging, similar to code testing workflows.
   - **Domain knowledge embedding**: Inject terminology and standards in professional scenarios (e.g., legal documents referencing specific statutes).
   - **Context management**: Dynamically truncate or retain conversation history to balance token consumption.

---

### **III. Application Scenarios**
1. **Content creation**: Generate structured content like marketing copy and academic reports (e.g., e-commerce livestream scripts need price anchors and interaction mechanisms).
2. **Professional efficiency**: Assist with legal compliance reviews, financial data analysis, and other high-accuracy tasks.
3. **Cross-modal generation**: Describe style parameters in image generation (e.g., "cyberpunk style, neon lighting").

---

### **IV. Current Status and Challenges**
- **Professionalization trend**: Prompt Engineer has become an emerging role, requiring both domain knowledge and model understanding.
- **Technical challenges**: Need to address model hallucinations, ethical risks (e.g., biased outputs), and long-context processing limitations.
- **Tool support**: Platforms like Volcengine Ark already provide full-lifecycle management tools for prompt development and evaluation.

Through systematic design, prompt engineering is driving AI from "tool" to "collaborative partner," becoming the core methodology for unlocking LLM potential.

## Prompts for Function Calling

Using an LLM to control drone flight essentially means using the LLM to generate code that controls the drone to complete various tasks. The simplest approach is to use prompts to invoke drone-related SDK functions.

### Role Usage
- system role definition: e.g., data scientist (can only call analysis APIs)
- user role permissions: submit raw data requests, call functions, etc.


### Function Description Templates



#### **Template 1: Simple Description**
```text
aw.takeoff() - Take off the drone.
aw.land() - Land the drone.
aw.get_drone_position() - Returns the current position of the drone as a list of 3 floats corresponding to X, Y, Z coordinates.

```
*Characteristics*:
- Simple and easy to understand
- Suitable for simple function calls


#### **Template 2: With Parameter Description**
```text
Function name: validate_user_registration
Purpose: Validate user registration information
Input parameters:
- username: string (4-20 characters, alphanumeric only)
- password: string (must contain uppercase letter, number, and special character)
- email: email address conforming to RFC 5322 standard
Return value:
- {"status": "success/error", "message": "detailed description"}
```
*Characteristics*:
- Parameters have expression constraints
- Return value structure is predefined

---

#### **Template 3: Scenario Application Template**
```text

Available function list:
1. Function name: query_customer_orders
   Purpose: Query customer order history
   Input:
   - customer_id: integer (required)
   - start_date: date format YYYY-MM-DD (default: last 30 days)
   Output:
   - JSON array containing order ID, amount, status
2. Function name: update_order_status
   Purpose: Update order logistics status
   Input:
   - order_id: integer (required)
   - new_status: enum ["shipped", "returning", "completed"]
   Output:
   - {"affected_rows": integer}
```
*Optimization points*:
- Standardized date format (avoids ambiguity)
- Status field enum restriction

---


#### **Error Handling Template**
```text
Function name: process_payment
Purpose: Credit card payment processing
Input:
- card_number: string (Luhn algorithm validated)
- amount: numeric (>= $0.01)
- currency: enum ["USD", "EUR"]
Error codes:
- 1001: Invalid card number
- 1002: Insufficient balance
- 1003: Currency mismatch
Return:
- Success: {"transaction_id": string}
- Failure: {"error_code": integer, "suggestion": resolution advice}
```
*Advantages*:
- Algorithm-level input parameter validation
- Error code to solution mapping


### Evolution of LLM Function Calling

Comparison of **Prompt-based**, **Function Call**, and **Model Control Protocol (MCP)** approaches

---

| **Feature**           | **Prompt-based**                                                                      | **Function Call**                                                              | **Model Control Protocol (MCP)**                                              |
|-----------------------|---------------------------------------------------------------------------------------|--------------------------------------------------------------------------------|-------------------------------------------------------------------------------|
| **Core Principle**    | Guides model output through natural language task descriptions.                       | Model generates JSON-format call requests; external tools execute and return results. | Standardized protocol supporting multi-model collaboration, dynamic task assignment, and real-time monitoring. |
| **Use Cases**         | Simple tasks (text generation, Q&A, translation).                                     | Complex tasks (database queries, API calls, computation).                       | Multi-model collaborative tasks (text + image generation + data analysis).     |
| **Flexibility**       | High: freely describe tasks in natural language.                                      | Medium: task scope limited to predefined functions.                             | High: supports dynamic task assignment and multi-model collaboration.          |
| **Controllability**   | Low: output may deviate from expectations, requires multiple prompt iterations.       | High: constrains output through structured parameters, reduces hallucinations.  | High: defines model behavior and interaction rules through config files.       |
| **Development Cost**  | Low: no extra tools needed, direct natural language interaction.                      | High: requires pre-defining functions and implementing logic.                   | High: requires mastering protocol specs and configuring toolchains.            |
| **Efficiency**        | Low: complex tasks need multiple interactions, high token consumption.                | High: complex tasks completed quickly via external tools.                       | Medium: multi-model collaboration requires more compute resources.             |
| **Standardization**   | Low: prompt design depends on developer experience.                                   | Medium: must follow function calling conventions.                               | High: unified interfaces and config files reduce integration difficulty.       |
| **Example**           | ```You are a translator, translate "Hello, world!" to Chinese```                      | ```{"function": "get_weather", "parameters": {"location": "Beijing"}}```       | ```models: [{name: gpt-4, role: text_generator}, {name: sd, role: image_generator}]``` |
| **Pros**              | High flexibility, low development cost.                                               | High controllability and efficiency.                                            | High extensibility and standardization.                                       |
| **Cons**              | Low controllability and efficiency.                                                   | High development cost, low flexibility.                                         | High complexity, high resource consumption.                                   |
| **Future Direction**  | Combine with reinforcement learning to auto-generate efficient prompts.               | Support dynamic function registration and real-time updates.                    | More vendors and developers joining the protocol ecosystem.                   |

---

### **Summary**
1. **Prompt-based**: Suitable for quick validation and simple tasks, but lower controllability and efficiency.
2. **Function Call**: Suitable for complex tasks with high controllability and efficiency, but higher development cost.
3. **MCP**: Suitable for multi-model collaborative tasks with high extensibility and standardization, but higher complexity and resource consumption.

Choose the appropriate approach based on specific task requirements, or combine multiple approaches for optimal results.

````python
import os
from openai import OpenAI

# Make sure you have set your API key
# Initialize OpenAI client
client = OpenAI(
    # Default endpoint URL
    base_url="https://api.intelligence.io.solutions/api/v1",
    api_key="YOUR_IO_API_KEY",
)
````

````python
chat_history = []
````

````python
def chat(prompt, chat_history):
    chat_history.append(
            {
                "role": "user",
                "content": prompt,
            }
    )
    completion = client.chat.completions.create(
        model="moonshotai/Kimi-K2.6",
        messages=chat_history, #chat_history[-10:0]
        temperature=0.1,
    )

    content = completion.choices[0].message.content

    # Add assistant reply, saving full history for multi-turn conversation
    chat_history.append(
            {
                "role": "assistant",
                "content": content,
            }
        )

    return content
````

````python
SYS_MSG="""
You are an assistant helping me with the AirSim drone simulator.
When I ask you to do something, you should provide the Python code needed to accomplish the task (using only AirSim and related functions), and explain what the code does.
You can only use the functions I have defined for you. You cannot assume or use any other functions you think might exist.
You can use simple Python library functions such as `math` and `numpy`.
"""
chat_history.append(
    {
        "role": "system",
        "content": SYS_MSG,
    }
)
````

````python
knowledge_prompt="""
Here are some functions you can use to command the drone.

aw.takeoff() - Take off the drone.
aw.land() - Land the drone.

Reply in the following example format:
```python
i=1  # output python code here
```
This code assigns a value.

You do not need to worry about importing aw; it is already declared in the environment.

"""

ret = chat(knowledge_prompt, chat_history)
ret
````

````python
# Parse python code
import re
def extract_python_code(content):
    """
    Extracts the python code from a response.
    :param content:
    :return:
    """
    code_block_regex = re.compile(r"```(.*?)```", re.DOTALL)
    code_blocks = code_block_regex.findall(content)
    if code_blocks:
        full_code = "\n".join(code_blocks)

        if full_code.startswith("python"):
            full_code = full_code[7:]

        return full_code
    else:
        return None
````
