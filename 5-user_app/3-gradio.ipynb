```python
Gradio is a Python library for creating interactive interfaces for machine learning models. With Gradio, developers can quickly build a visual, easy-to-use web interface for their models without writing any frontend web code.

Main Features of Gradio:

1. Multiple input and output types: Gradio supports various input types (such as text, images, audio, etc.) and output types (such as text, images, HTML, etc.), and you can define the relationships between these inputs/outputs and processing functions directly in a Python script.

2. Rapid creation and deployment: Gradio can quickly create an interactive interface and generate a URL through its built-in sharing feature, allowing users to open this URL in a browser to interact with the model.

3. Easy to use: Gradio's design philosophy is "no-code" — developers only need a few lines of code to create an interactive preview interface for their models.

pip install gradio

# Tsinghua mirror source
pip install -i https://pypi.tuna.tsinghua.edu.cn/simple gradio
```

```python
import gradio as gr

# Define a simple function
def greet(name):
    return f"Hello, {name}!"

# Create Gradio interface
with gr.Blocks() as demo:
    # Add title
    gr.Markdown("### Simple Greeting Program")

    # Input component: text box
    name_input = gr.Textbox(label="Enter your name")

    # Output component: text box
    output_text = gr.Textbox(label="Greeting result")

    # Button to trigger processing
    submit_button = gr.Button("Submit")

    # Bind button click event to function
    submit_button.click(greet, inputs=name_input, outputs=output_text)

# Launch Gradio app
demo.launch()
```

# Calculator

```python
import gradio as gr

# Define a simple function
def add_numbers(num1, num2):
    return num1 + num2

# Create Gradio interface
with gr.Blocks() as demo:
    # Add title
    gr.Markdown("## Simple Addition Calculator")

    # Input components
    num1_input = gr.Number(label="Enter the first number", value=0)
    num2_input = gr.Number(label="Enter the second number", value=0)

    # Output component
    output = gr.Number(label="Result")

    # Button bound to function
    add_button = gr.Button("Calculate")
    add_button.click(add_numbers, inputs=[num1_input, num2_input], outputs=output)

# Launch interface
demo.launch()
```

# Text


```python
import gradio as gr

# Define a text generation function
def generate_response(prompt):
    return f"You entered: {prompt}"

# Create Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Text Generator")

    with gr.Row():
        input_text = gr.Textbox(label="Enter your question", lines=3)
        output_text = gr.Textbox(label="Generated result", lines=3)

    submit_button = gr.Button("Generate")
    submit_button.click(generate_response, inputs=input_text, outputs=output_text)

# Launch interface
demo.launch()
```

```python
import gradio as gr
scores = []
def track_score(score):
    scores.append(score)
    # Return top 3 scores
    top_scores = sorted(scores, reverse=True)[:3]
    return top_scores
demo = gr.Interface(
    track_score,
    gr.Number(label="Score"),
    gr.JSON(label="Top Scores")
)
demo.launch()
```

```python
import gradio as gr
from PIL import Image, ImageOps

# Define function to convert image to grayscale
def convert_to_grayscale(image):
    # Use PIL library to convert image to grayscale
    grayscale_image = ImageOps.grayscale(image)
    return grayscale_image

# Create Gradio interface
with gr.Blocks() as demo:
    with gr.Row():
        input_image = gr.Image(label="Upload image", type="pil")  # Input image component
        output_image = gr.Image(label="Grayscale image", type="pil")  # Output image component
    submit_button = gr.Button("Convert to grayscale")  # Submit button

    # Bind button click event to processing function
    submit_button.click(convert_to_grayscale, inputs=input_image, outputs=output_image)

# Launch Gradio app
demo.launch(share=True)
```

```python
import gradio as gr

def show_info(name, age, subscribe, gender):
    result = f"""
    Name: {name}
    Age: {age}
    Subscribed: {'Yes' if subscribe else 'No'}
    Gender: {gender}
    """
    return result

demo = gr.Interface(
    fn=show_info,
    inputs=[
        gr.Textbox(label="Name"),
        gr.Slider(0, 100, label="Age"),
        gr.Checkbox(label="Subscribe"),
        gr.Radio(["Male", "Female", "Other"], label="Gender")
    ],
    outputs=gr.Textbox(label="User Info")
)

demo.launch()
```
