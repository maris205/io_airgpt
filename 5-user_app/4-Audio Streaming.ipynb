# 5.4 Voice-Controlled Drone

```python
import gradio as gr
from pydub import AudioSegment
import os

# Make sure ffmpeg is installed
# If not installed, run the following commands:
# pip install pydub
# And download and install ffmpeg: https://ffmpeg.org/

def save_audio_as_mp3(audio):
    # audio is a tuple containing sample_rate and audio_data
    if isinstance(audio, str):
        # If it's a file path, load the audio file
        audio_segment = AudioSegment.from_file(audio)
    else:
        # If it's a NumPy array, convert to PyDub AudioSegment object
        sample_rate, audio_data = audio
        audio_segment = AudioSegment(
            audio_data.tobytes(),  # Audio data
            frame_rate=sample_rate,  # Sample rate
            sample_width=audio_data.dtype.itemsize,  # Sample width
            channels=1  # Mono
        )

    # Save as MP3 file
    output_file = "output2.mp3"
    audio_segment.export(output_file, format="mp3")

    return f"Audio saved as {output_file}"

# Create Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("### Real-time Audio Recording and Save as MP3 Example")

    # Use gr.Audio component, supports recording from microphone or uploading audio files by default
    audio_input = gr.Audio(label="Record audio or upload file", type="numpy")

    # Output component to display processing result
    output_text = gr.Textbox(label="Processing result")

    # Button to trigger processing
    submit_button = gr.Button("Save as MP3")

    # Bind button click event to processing function
    submit_button.click(save_audio_as_mp3, inputs=audio_input, outputs=output_text)

# Launch Gradio app
demo.launch()
```
