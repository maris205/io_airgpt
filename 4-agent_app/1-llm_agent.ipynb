# 4.1 Introduction to LLM Agents

Even a flashy cobra maneuver performed by a fighter jet, from a task planning perspective, is essentially composed of basic OODA (Observation - Orientation - Decision - Action) loops.

Anything that can complete this loop can be called an Agent. For example, a model that helps you book tickets -- one that perceives your command and then executes the booking operation on a website -- is also an Agent.

**LLM Agents** are intelligent agent systems built on Large Language Models (LLMs) that can autonomously analyze, plan, and execute complex tasks, breaking through the limitations of traditional AI that relies on explicit instructions. Their core strength lies in combining the LLM's semantic understanding capabilities with tool calling, dynamic decision-making, and other modules to achieve a complete closed-loop from perception to action.

"Observation" refers to acquiring environmental data, typically through sensors such as visual cameras, position radar, etc. Simulation systems have corresponding APIs for these.

"Action" is generally divided into basic movements and complex behaviors. For example, a drone's basic movements include attitude control (pitch, roll, yaw) and position control (forward/backward, left/right, up/down) -- that is, 6 degrees of freedom. Based on these basic operations, complex behaviors like flips can be composed. The AirSim system provides basic motion control as well as complex behavior APIs like path planning.

"Perception" is the understanding of the environment built upon "Observation," such as YOLO recognition in machine learning, and the currently popular multimodal perception models.

"Planning" is the decision-making and reasoning process, such as decision planning based on reinforcement learning models, or decisions based on large language models.

LLM-based drone control essentially means: based on environmental observation and perception, using an LLM for decision-making, calling action APIs to complete specific tasks.

#### Core Capabilities
1. **Reasoning and Inference**
- Through Chain-of-Thought (CoT), Tree-of-Thought (ToT) and other methods to decompose tasks. For example, breaking down "create a data analysis report" into steps like data collection, analysis, and visualization.
- Dynamically adjusting execution paths, generating alternative solutions when failures occur (e.g., retrying, switching tools).

2. **Multimodal Processing**
- Processing text, images, and other data types, supporting natural language interaction and tool/API invocation. For example, in a visual scene, an Agent can parse CT images to generate diagnostic reports.

3. **Context Memory**
- Maintaining long-term memory and task state. For example, an Agent can remember previous interaction history to maintain policy continuity.

---

#### Key Technologies
1. **Tool Calling**
- **Static calling**: Using LLM to invoke predefined tools (search engines, maps, etc. with fixed parameters).
- **Dynamic calling**: Leveraging reinforcement learning for adaptive tool selection (data -> statistical analysis -> visualization tool).
- **Execution environment**: API interfaces and sandboxed execution environments (for Agent code execution).

2. **Multi-Agent Collaboration**
- Multiple Agents supporting parallel task processing. For example, multiple Agents collaborating on monitoring, analysis, and report generation.

### LLM Agents vs. Workflows

LLM Agents and Workflows are both approaches to AI task execution, but they differ significantly in design philosophy, technical implementation, and application scenarios.

Here are the key differences:

---

#### **1. Definitions**
- **Workflow**
A workflow is a **predefined and structured set of task execution steps**, emphasizing process standardization and repeatability. For example, a document approval process like "submit -> review -> approve" is a workflow.
- **Features**: Fixed path, deterministic, executed step by step.


<img src="img/workflow.png" width='640px' />

- **LLM Agent**
An Agent is an **autonomous entity that can perceive its environment, make decisions, and execute actions**, built on large language models (LLMs) to implement intelligent task processing. For example, an Agent can use search tools and call APIs to complete tasks.
- **Features**: Autonomous, goal-driven, adaptive.

<img src="img/drone_agent.png" width='640px' />


---

#### **2. Execution Logic**
- **Workflow**
- **Deterministic**: Process paths are fixed -- after submission comes testing, then deployment.
- **Controlled**: Process is hardcoded, executing operations step by step.

- **LLM Agent**
- **Dynamic**: Makes decisions based on environmental context, adapting responses to content.
- **Uncertain**: Agents have autonomy and may encounter errors when calling tools.

---

#### **3. Technical Implementation**
- **Workflow**
- **Tools**: BPMN, Airflow and other process management tools, with conditional branching.
- **Complexity**: Low, based on process graphs or rule-based task execution.

- **LLM Agent**
- **Tech stack**: LLM, reinforcement learning, tool calling (search APIs), and memory.
- **Complexity**: High, with multimodal support.

---

#### **4. Application Scenarios**
- **Workflow use cases**
- **Highly standardized, process-oriented tasks**: Approvals, deployments.
- **Repetitive tasks**: Batch processing, scheduled jobs.

- **LLM Agent use cases**
- **Dynamic environments**: Autonomous navigation (tool use), obstacle avoidance.
- **Complex reasoning scenarios**: Data analysis and solution generation, evaluation.

---

#### **5. Hybrid Approaches**
- **Agents within Workflows**
Agents can be embedded within workflow steps. For example, in a deployment workflow, an Agent can make decisions based on test results.

- **Multi-Agent Workflows**
Multiple Agents can collaborate through workflows. For example, a "generate -> review -> publish" process where different Agents handle each step.

---

### **Summary**
- **Workflows**: Best for deterministic, standardized task execution with fixed paths.
- **Agents**: Best for non-deterministic scenarios requiring reasoning and data analysis.
- **Hybrid**: Combining both approaches leverages the strengths of each.

### Popular LLM Agent Frameworks

LLM Agents achieve intelligent task processing and execution through integration with various tools. Here are the major frameworks and their characteristics:

---

#### 1. **LangChain**
- **Overview**: A comprehensive Agent framework supporting extensive tool integration.
- **Features**:
  - Provides rich integrations (200+ tools), including indexing, data processing, and execution environments.
  - Supports multiple LLMs (GPT-4, Llama-2) and multimodal inputs.
- **Use cases**: Scenarios requiring rapid integration of external tools, RAG, data analysis.
- **Drawbacks**: Complex abstractions, steep learning curve, difficult debugging.

---

#### 2. **Microsoft AutoGen**
- **Overview**: A multi-agent framework designed for AI collaboration and conversation.
- **Features**:
  - Supports defining custom agents and task allocation.
  - Multi-agent collaboration, e.g., "query -> analyze -> reply" pipelines.
- **Use cases**: Multi-agent scenarios (collaborative coding, debate simulation).
- **Drawbacks**: Limited multimodal task and tool support.

---

#### 3. **Dify**
- **Overview**: A rapid development platform for LLM applications.
- **Features**:
  - Built-in 50+ tools (Google Search, DALL-E), supports RAG and ReAct patterns.
  - Provides visual interface for quickly building and deploying applications.
- **Use cases**: Rapid LLM application development, chatbots, content generation.
- **Drawbacks**: Limited scenario support, restricted API calling capabilities.

---

#### 4. **MetaGPT**
- **Overview**: A multi-agent framework that standardizes task processes through coding.
- **Features**:
  - Uses SOP (Standard Operating Procedures) to decompose tasks.
  - Supports role-based collaboration, e.g., "design -> code -> test" pipelines.
- **Use cases**: Complex collaborative tasks (project management, process automation).
- **Drawbacks**: Limited support for non-coding tasks (image processing).


### 5. **SmolAgents**
**Overview**: Hugging Face's open-source Agent framework focused on **code-first** task execution, supporting rapid creation of highly capable agents.

**Features**:
- **Code-first**: Generates and executes Python code to complete tasks, reducing intermediate steps (calculations, searches).
- **Rich tool ecosystem**: Built-in 50+ tools (search, image APIs), supports custom tools and Hugging Face Hub integration.
- **Lightweight**: Create an Agent in just 3 lines of code. Supports Transformers, Ollama, and OpenAI-compatible APIs.

---

#### 6. **OmAgent**
- **Overview**: A multimodal agent framework for embodied AI.
- **Features**:
  - Supports various sensors (cameras, LiDAR) for perception.
  - Provides usage examples (video understanding, task planning) and documentation.
- **Use cases**: Robotics, monitoring systems (surveillance, inspection).
- **Advantages**: Low barrier to entry, supports multiple reasoning patterns (ReAct, DnC).

---


### Summary
- **Full-featured**: LangChain and Phidata are suitable for complex scenarios requiring extensive tool integration.
- **Multi-Agent**: AutoGen and MetaGPT excel in collaborative task decomposition.
- **Embodied AI**: OmAgent and SmolAgents are ideal for physical-world applications.
- **Rapid prototyping**: Dify and Lagent offer low barriers to entry for quick deployment.
