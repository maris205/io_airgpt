## Multimodal LLMs

### What are Multimodal LLMs?
**Multimodal LLMs** are models capable of processing and understanding multiple data types (text, images, audio, video, etc.). Through alignment, fusion, and generation technologies, they bridge different data modalities to achieve comprehensive scene understanding. Multimodal LLMs offer the following advantages:
1. **Cross-modal understanding**: Combining information from multiple modalities improves task accuracy. For example, analyzing context through both images and text.
2. **Generation capabilities**: Supporting text-to-image, video, or audio generation (e.g., DALL-E, Stable Diffusion).
3. **Transfer learning**: Pre-training on multimodal data enables strong performance across many tasks including classification, detection, and captioning.

#### Key Technologies
- **Alignment and Fusion**: Using encoders (like CLIP) to map different data types into a shared representation space.
- **Pre-training**: Using Mixture of Experts (MoE), Transformers, etc. for multi-task pre-training.
- **Generation and Reasoning**: Leveraging attention mechanisms for high-quality content generation, and Chain-of-Thought (CoT) for complex reasoning.

#### Application Scenarios
1. **Multimodal assistants**: LLMs that can analyze data, generate charts, and answer questions about images.
2. **Autonomous systems**: Models like DeepSeek enabling natural language control, safe and reliable operation.
3. **Edge deployment**: Large LLMs supporting multimodal data processing on edge devices with fast response times.
4. **Research**: Meta ImageBind, Emu3, etc. supporting cross-modal understanding and generation.

### Key Technologies and Training
1. **Visual Encoders**
- **OCR-Free**: Using CLIP-like models for direct scene understanding.
- **OCR-Based**: LayoutLMv3 etc. for text-rich image parsing (documents, forms).

2. **Alignment and Decoding**
- **Connection modules**: Using attention mechanisms to bridge vision and language models.
- **Training stages**:
  - **Alignment pre-training**: Learning cross-modal correspondences (CLIP-style contrastive learning).
  - **Instruction tuning**: Task-specific fine-tuning for improved reasoning.

```python
import base64

# Define image encoding function
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

local_image_path = "img/p-4-1-s.png"

# Image input:
response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What objects are in this image?"},
                {
                    "type": "image_url",
                    "image_url": {
                        # Use Base64-encoded local image, note img/png or img/jpg format
                        "url": f"data:image/png;base64,{encode_image(local_image_path)}"
                    }
                },
            ],
        }
    ],
)

print(response.choices[0])
```

```python
# Test OpenCV image reading
import cv2
import base64

def cv2_to_base64(image, format='.png'):
    """Convert an OpenCV numpy array image to a Base64 string"""
    # Encode as byte stream
    success, buffer = cv2.imencode(format, image)
    if not success:
        raise ValueError("Image encoding failed, please check the format parameter")

    # Convert to Base64
    img_bytes = buffer.tobytes()
    return base64.b64encode(img_bytes).decode('utf-8')

def cv_read(image_path):
    # Read image as BGR format numpy array
    image = cv2.imread(image_path)
    # BGR to RGB conversion (if needed)
    # Note: Most APIs (like GPT-4o) handle this automatically
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    base64_str = cv2_to_base64(rgb_image, format='.png')  # or '.jpg'
    return base64_str


local_image_path = "img/p-4-1-s.png"

# Image input:
response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Describe this image"},
                {
                    "type": "image_url",
                    "image_url": {
                        # Use Base64-encoded local image, note img/png or img/jpg format
                        "url": f"data:image/png;base64,{cv_read(local_image_path)}"
                    }
                },
            ],
        }
    ],
)

print(response.choices[0])
```
